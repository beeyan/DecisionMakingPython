```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

### ランチェスターの第1戦略

```python
# 解きたい微分方程式の関数
def odeLan1(t, z, cx, cy):
    dx = -cy
    dy = -cx
    return [dx, dy]
```

```python
# 初期条件の兵隊の数
x0, y0 = 5.0, 7.0
```

```python
# case.1: 戦力が同等のとき
Tend = 5.0 
cx, cy = 1.0, 1.0
sol = solve_ivp(odeLan1, t_span=[0, Tend], y0=[x0, y0], args=(cx, cy,), dense_output=True)
tc1 = np.linspace(0, Tend, 100)  
yc1 = sol.sol(tc1)
```

```python
# case.2: cx > cy : Xの方が強いケース

Tend = 2.5
cx, cy = 3.0, 1.0 
sol = solve_ivp(odeLan1, t_span=[0, Tend], y0=[x0, y0], args=(cx, cy,), dense_output=True)
tc2 = np.linspace(0, Tend, 100)  
yc2 = sol.sol(tc2)
```

```python
# case.3: cx > cy : Xの方が少しだけ強いケース

Tend = 10
cx, cy = 1.01, 1.0 
sol = solve_ivp(odeLan1, t_span=[0, Tend], y0=[x0, y0], args=(cx, cy,), dense_output=True)
tc3 = np.linspace(0, Tend, 100)  
yc3 = sol.sol(tc2)
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(14,4))

ax[0].plot(tc1, yc1[0].T, label='x')
ax[0].plot(tc1, yc1[1].T, label='y')
ax[0].grid()
ax[0].legend()
ax[0].set_xlabel('time')
ax[0].set_ylabel('number of forces')

ax[1].plot(tc2, yc2[0].T, label='x')
ax[1].plot(tc2, yc2[1].T, label='y')
ax[1].grid()
ax[1].legend()
ax[1].set_xlabel('time')
ax[1].set_ylabel('number of forces')

ax[2].plot(tc3, yc3[0].T, label='x')
ax[2].plot(tc3, yc3[1].T, label='y')
ax[2].grid()
ax[2].legend()
ax[2].set_xlabel('time')
ax[2].set_ylabel('number of forces')

plt.show()
```

### ランチェスターの第2法則

```python
def odeLan2(t, z, cx, cy):
    dx = -cy*z[1]
    dy = -cx*z[0]
    return [dx, dy]
```

```python
x0, y0 = 5.0, 7.0
```

```python
# case.1: 戦力が同等のとき

Tend = 0.9
cx, cy = 1.0, 1.0
sol = solve_ivp(odeLan2, t_span=[0, Tend], y0=[x0, y0], args=(cx, cy,), dense_output=True)
tc4 = np.linspace(0, Tend, 100)  
yc4 = sol.sol(tc4)
```

```python
# case.3: cx > cy : Xの方が2倍強いケース
Tend = 0.9
cx, cy = 2.0, 1.0
sol = solve_ivp(odeLan2, t_span=[0, Tend], y0=[x0, y0], args=(cx, cy,), dense_output=True)
tc6 = np.linspace(0, Tend, 100)  
yc6 = sol.sol(tc6)
```

### リチャードソンモデル

```python
def odeRichard(t, z, k, l, alpha, beta, g, h):
    dx = k*z[1] - alpha*z[0] + g
    dy = l*z[0] - beta*z[1] + h
    return [dx, dy]
```

```python
# case.1 均衡点に収束する場合
k, l = 1, 1
alpha, beta = 2, 2
g, h = 2, 3

x0, y0 = 10.0, -4.0
Tend = 7.0

sol = solve_ivp(odeRichard, t_span=[0, Tend], y0=[x0, y0], args=(k, l, alpha, beta, g, h,), dense_output=True)
tc1 = np.linspace(0, Tend, 100)  
yc1 = sol.sol(tc1)
```

```python
# case.2 均衡点に収束せず、軍拡が進む場合
k, l = 2, 2
alpha, beta = 4, 1
g, h = 2, 3

x0, y0 = 10.0, -4.0
Tend = 7.0

sol = solve_ivp(odeRichard, t_span=[0, Tend], y0=[x0, y0], args=(k, l, alpha, beta, g, h,), dense_output=True)
tc2 = np.linspace(0, Tend, 100)  
yc2 = sol.sol(tc1)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(14,8))

ax[0][0].plot(tc1, yc1[0].T, label='x')
ax[0][0].plot(tc1, yc1[1].T, label='y')
ax[0][0].grid()
ax[0][0].legend()
ax[0][0].set_xlabel('time')
ax[0][0].set_ylabel('amount of armaments')

ax[0][1].plot(yc1[0].T, yc1[1].T, label='x-y')
ax[0][1].grid()
ax[0][1].legend()
ax[0][1].set_xlabel('x')
ax[0][1].set_ylabel('y')

ax[1][0].plot(tc2, yc2[0].T, label='x')
ax[1][0].plot(tc2, yc2[1].T, label='y')
ax[1][0].grid()
ax[1][0].legend()
ax[1][0].set_xlabel('time')
ax[1][0].set_ylabel('amount of armaments')

ax[1][1].plot(yc2[0].T, yc2[1].T, label='x-y')
ax[1][1].grid()
ax[1][1].legend()
ax[1][1].set_xlabel('x')
ax[1][1].set_ylabel('y')

#plt.savefig('fig_DE_Richardson_01.pdf', bbox_inches='tight')
plt.show()
```
